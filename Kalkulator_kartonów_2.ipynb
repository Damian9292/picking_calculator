```python
import tkinter as tk
from tkinter import messagebox
import os
import sys

def cartoons_calculator():
    global global_start  # Zmienna globalna, aby była dostępna w innych funkcjach

    def add_to_summary():
        try:
            x = float(entry_x.get().replace(',', '.'))
            y = float(entry_y.get().replace(',', '.'))
            z = float(entry_z.get().replace(',', '.'))
        except ValueError:
            messagebox.showerror("Błąd", "Błędny format danych, proszę poprawić :)")
            return

        if not summary_data:  # Jeśli to pierwsza iteracja
            wynik = global_start - (x * y * z)
            preparowano = x * y * z
        else:  # Kolejne iteracje
            wynik = summary_data[-1][4] - (x * y * z)
            preparowano = x * y * z

        summary_data.append((x, y, z, preparowano, wynik))
        update_summary()

        entry_x.delete(0, tk.END)
        entry_y.delete(0, tk.END)
        entry_z.delete(0, tk.END)

    def update_summary():
        text_widget.config(state=tk.NORMAL)
        text_widget.delete(1.0, tk.END)
        text_widget.insert(tk.END, "{:<20} | {:<20} | {:<25} | {:<25} | {:<25}\n".format('Ilość osób', 'Ilość godzin', 'Średnia kartony/h', 'Spreparowano', 'Pozostały global'))
        text_widget.insert(tk.END, "-" * 120 + "\n")
        for data in summary_data:
            line_text = "{:<20} | {:<20} | {:<25} | {:<25} | {:<25}".format(*data)
            text_widget.insert(tk.END, line_text + "\n")
            text_widget.insert(tk.END, "-" * 120 + "\n")
        text_widget.config(state=tk.DISABLED)

    def start_calculation():
        global global_start  # Zmienna globalna
        try:
            global_start = float(entry_global.get().replace(',', '.'))
        except ValueError:
            messagebox.showerror("Błąd", "Błędny format danych, proszę poprawić :)")
            return

        # Blokujemy pole globalu i przycisk, odblokowujemy pozostałe pola
        entry_global.config(state=tk.DISABLED)
        start_button.config(state=tk.DISABLED)
        entry_x.config(state=tk.NORMAL)
        entry_y.config(state=tk.NORMAL)
        entry_z.config(state=tk.NORMAL)
        add_button.config(state=tk.NORMAL)
        entry_x.focus()

    # Funkcja pomocnicza do uzyskania ścieżki pliku
    def resource_path(relative_path):
        """Funkcja zwracająca absolutną ścieżkę pliku, niezależnie od tego, czy aplikacja jest uruchamiana lokalnie, czy jako plik wykonywalny"""
        try:
            # Próbujemy uzyskać ścieżkę do pliku wbudowanego przez PyInstaller
            base_path = getattr(sys, '_MEIPASS', os.path.dirname(os.path.abspath(__file__)))
            return os.path.join(base_path, relative_path)
        except Exception as e:
            print(f"Error accessing file path: {e}")
            return None

    # Tworzenie głównego okna
    root = tk.Tk()
    root.state('zoomed')  # Pełny ekran
    root.title("Kalkulator Kartonów")

    summary_data = []

    # Lewy górny róg - Podsumowanie
    summary_frame = tk.Frame(root, bg='lightgrey')
    summary_frame.pack(side=tk.LEFT, anchor='nw', padx=20, pady=20, fill=tk.BOTH, expand=True)

    text_widget = tk.Text(summary_frame, font=("Courier", 12), width=100, height=40, bg='white')
    text_widget.pack(fill=tk.BOTH, expand=True)
    text_widget.config(state=tk.DISABLED)

    # Prawy górny róg - Dane wejściowe
    input_frame = tk.Frame(root, bg='lightgrey')
    input_frame.pack(side=tk.RIGHT, anchor='ne', padx=20, pady=20)

    tk.Label(input_frame, text="Global do preparacji:", font=("Arial", 14), bg='lightgrey').grid(row=0, column=0, sticky='e', pady=5)
    entry_global = tk.Entry(input_frame, font=("Arial", 14))
    entry_global.grid(row=0, column=1, pady=5)

    tk.Label(input_frame, text="Ilość osób:", font=("Arial", 14), bg='lightgrey').grid(row=1, column=0, sticky='e', pady=5)
    entry_x = tk.Entry(input_frame, font=("Arial", 14), state=tk.DISABLED)
    entry_x.grid(row=1, column=1, pady=5)

    tk.Label(input_frame, text="Ilość godzin:", font=("Arial", 14), bg='lightgrey').grid(row=2, column=0, sticky='e', pady=5)
    entry_y = tk.Entry(input_frame, font=("Arial", 14), state=tk.DISABLED)
    entry_y.grid(row=2, column=1, pady=5)

    tk.Label(input_frame, text="Średnia kartony/h:", font=("Arial", 14), bg='lightgrey').grid(row=3, column=0, sticky='e', pady=5)
    entry_z = tk.Entry(input_frame, font=("Arial", 14), state=tk.DISABLED)
    entry_z.grid(row=3, column=1, pady=5)

    # Przyciski
    start_button = tk.Button(input_frame, text="Rozpocznij", font=("Arial", 14), command=start_calculation)
    start_button.grid(row=4, column=0, columnspan=2, pady=10)

    add_button = tk.Button(input_frame, text="Dodaj", font=("Arial", 14), command=add_to_summary, state=tk.DISABLED)
    add_button.grid(row=5, column=0, columnspan=2, pady=10)

    root.mainloop()

cartoons_calculator()
```
